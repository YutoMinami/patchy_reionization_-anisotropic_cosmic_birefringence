# 22b visualize natural-unit budget

`21` の natural-unit normalization を使って描き直した `22` の budget を確認する notebook。


```python
from pathlib import Path
import pandas as pd
from IPython.display import display, Image, Markdown

ROOT = Path('/home/yminami/workdir/CosmicBirefringence/PatchyReionization')
OUT = ROOT / 'results' / '22-natural-unit-budget'
CSV = OUT / 'natural_unit_budget_scan.csv'
df = pd.read_csv(CSV)
display(Markdown(f'**Output dir:** `{OUT}`'))
display(df.head())
```

```python
summary = []
for g, sub in df.groupby('g_GeV_inv'):
    frac = sub['D_alphaalpha_tau'] / sub['D_alphaalpha_limit']
    i = frac.idxmax()
    row = sub.loc[i]
    summary.append({
        'g_GeV_inv': g,
        'delta_tau_at_max': row['delta_tau'],
        'L_at_max': int(row['L']),
        'max_patchy_to_limit': frac.loc[i],
        'A_tau_physical': row['A_tau_physical'],
    })
display(pd.DataFrame(summary))
```

```python
for name in [
    'deltatau_variation_dtautau.png',
    'physical_patchy_dalpha_vs_limit_g_1.4e-12.png',
    'physical_total_dalpha_vs_limit_g_1.4e-12.png',
    'physical_patchy_dalpha_vs_limit_g_4.0e-12.png',
    'physical_total_dalpha_vs_limit_g_4.0e-12.png',
]:
    display(Markdown(f'## {name}'))
    display(Image(filename=str(OUT / name)))
```
