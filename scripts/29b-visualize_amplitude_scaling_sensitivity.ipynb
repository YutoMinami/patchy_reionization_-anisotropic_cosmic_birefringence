# 29: Amplitude-Scaling Sensitivity

`D_{\rm peak} \propto (A/A_{\rm fid})^{p_{\rm amp}} (b/b_{\rm fid})^{p_b}` を仮定したとき、current parameter range でどれだけ boost を稼げるかを見る notebook。

```python
from pathlib import Path
import pandas as pd
import matplotlib.pyplot as plt

repo = Path.cwd().resolve().parent if Path.cwd().name == 'scripts' else Path('/home/yminami/workdir/CosmicBirefringence/PatchyReionization')
outdir = repo / 'results/29-amplitude-scaling-sensitivity'
df = pd.read_csv(outdir / 'amplitude_scaling_sensitivity.csv')
df
```

```python
pivot = df.pivot(index='p_amp', columns='p_b', values='max_boost')
pivot
```

```python
fig, ax = plt.subplots(figsize=(6, 4.8))
im = ax.imshow(pivot.values, origin='lower', aspect='auto')
ax.set_xticks(range(len(pivot.columns)), [str(x) for x in pivot.columns])
ax.set_yticks(range(len(pivot.index)), [str(x) for x in pivot.index])
ax.set_xlabel('p_b')
ax.set_ylabel('p_amp')
ax.set_title('Maximum boost within current (tau, Delta y, b) range')
for i in range(pivot.shape[0]):
    for j in range(pivot.shape[1]):
        ax.text(j, i, f'{pivot.values[i,j]:.1f}', ha='center', va='center', color='w')
fig.colorbar(im, ax=ax, label='max boost')
plt.show()
```

```python
req = pd.read_csv(repo / 'results/28-required-template-boost/required_template_boost.csv')
req1 = req[req['target_frac'] == 1e-2][['g_GeV_inv','required_boost']]
req1
```

```python
fig, ax = plt.subplots(figsize=(7, 4.5))
for p_amp, sub in df.groupby('p_amp'):
    ax.plot(sub['p_b'], sub['max_boost'], marker='o', label=f'p_amp = {p_amp:g}')
for _, row in req1.iterrows():
    ax.axhline(row['required_boost'], linestyle='--', alpha=0.7, label=f"1% req, g={row['g_GeV_inv']:.1e}")
ax.set_xlabel('p_b')
ax.set_ylabel('max boost')
ax.set_title('Can steeper scaling reach 1% of the limit?')
ax.grid(True, alpha=0.3)
ax.legend(fontsize=9)
plt.show()
```
