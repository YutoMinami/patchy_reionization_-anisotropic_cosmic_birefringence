# 18b visualize Delta tau variation budget

peak 位置はほぼ固定し、`\Delta\tau` variation だけで `C_L^{\tau\tau}` と birefringence budget を見る notebook です。

```python
from pathlib import Path
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.image as mpimg

ROOT = Path('/home/yminami/workdir/CosmicBirefringence/PatchyReionization')
RESULT = ROOT / 'results/18-deltatau-variation-budget'
df = pd.read_csv(RESULT / 'deltatau_budget_scan.csv')
df.head()
```

## Numeric setup

```python
cols = ['mass_eV','A_tau_eff','tau','delta_y','Rbar_Mpc','siglnR','Reff_Mpc','z_re','L_peak_surrogate','delta_tau_scale']
df[cols].drop_duplicates().sort_values('delta_tau_scale')
```

## $D_L^{\tau\tau}$ variation

```python
plt.figure(figsize=(8,5.5))
plt.imshow(mpimg.imread(RESULT / 'deltatau_variation_ctautau.png'))
plt.axis('off')
plt.tight_layout()
plt.show()
```

## Patchy-induced $C_L^{\alpha\alpha}$ vs limit

```python
plt.figure(figsize=(8,5.5))
plt.imshow(mpimg.imread(RESULT / 'patchy_alpha_vs_limit.png'))
plt.axis('off')
plt.tight_layout()
plt.show()
```

## Total benchmark $C_L^{\alpha\alpha}$ vs limit

```python
plt.figure(figsize=(8,5.5))
plt.imshow(mpimg.imread(RESULT / 'total_alpha_vs_limit.png'))
plt.axis('off')
plt.tight_layout()
plt.show()
```

## Reading guide

- 最初の図では、peak 位置をほぼ固定したまま高さだけが `\Delta\tau` で変わっているかを見る。
- 2枚目では patchy 側だけでどこまで limit に近づくかを見る。
- 3枚目では genuine term を一定割合だけ入れたときに total budget がどうなるかを見る。
