# 30: Best-Case Scaled Budget

`27b` の current best budget に `29` の最大 boost を掛けたとき、current parameter range の中で observable fraction がどこまで上がるかを見る notebook。

```python
from pathlib import Path
import pandas as pd
import matplotlib.pyplot as plt

repo = Path.cwd().resolve().parent if Path.cwd().name == 'scripts' else Path('/home/yminami/workdir/CosmicBirefringence/PatchyReionization')
outdir = repo / 'results/30-bestcase-scaled-budget'
df = pd.read_csv(outdir / 'bestcase_scaled_budget.csv')
df.head()
```

```python
summary = pd.read_csv(outdir / 'bestcase_scaled_budget_summary.csv')
summary
```

```python
fig, ax = plt.subplots(figsize=(7, 4.5))
for g, sub in df.groupby('g_GeV_inv'):
    sub = sub.sort_values(['p_amp','p_b'])
    labels = [f"({int(a)},{int(b)})" for a, b in zip(sub['p_amp'], sub['p_b'])]
    ax.plot(labels, sub['bestcase_frac_eff'], marker='o', label=f'g = {g:.1e} GeV^-1')
ax.axhline(1e-2, linestyle='--', color='gray', alpha=0.7, label='1% of limit')
ax.axhline(1e-1, linestyle=':', color='gray', alpha=0.7, label='10% of limit')
ax.set_yscale('log')
ax.set_xlabel('(p_amp, p_b) scaling')
ax.set_ylabel('Best-case fraction of anisotropic-CB limit')
ax.set_title('Best-case budget within the current surrogate parameter range')
ax.grid(True, which='both', alpha=0.3)
ax.legend(fontsize=9)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
