# 16b visualize Dvorkin-Smith figure reproduction

Fig.16 / Fig.18 風の surrogate 再現チェックを人間の目で確認する notebook です。

```python
from pathlib import Path
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.image as mpimg

ROOT = Path('/home/yminami/workdir/CosmicBirefringence/PatchyReionization')
PAPERS = ROOT / 'papers'
RESULT = ROOT / 'results/16-ds-figure-check'
summary = pd.read_csv(RESULT / 'figure_reproduction_summary.csv')
summary
```

## Paper reference pages

```python
fig, axes = plt.subplots(1, 2, figsize=(12, 7))
axes[0].imshow(mpimg.imread(PAPERS / 'p15.png'))
axes[0].set_title('paper p15')
axes[0].axis('off')
axes[1].imshow(mpimg.imread(PAPERS / 'p16.png'))
axes[1].set_title('paper p16')
axes[1].axis('off')
plt.tight_layout()
plt.show()
```

## Our Fig.16-like surrogate

```python
img = mpimg.imread(RESULT / 'fig16_like_surrogate.png')
plt.figure(figsize=(8,5.5))
plt.imshow(img)
plt.axis('off')
plt.tight_layout()
plt.show()
```

## Our Fig.18-like surrogate
