# 28: Required Template Boost

`27b` の best point を 1%, 10%, 100% of limit へ押し上げるのに必要な template-normalization boost を確認する notebook。

```python
from pathlib import Path
import pandas as pd
import matplotlib.pyplot as plt

repo = Path.cwd().resolve().parent if Path.cwd().name == 'scripts' else Path('/home/yminami/workdir/CosmicBirefringence/PatchyReionization')
outdir = repo / 'results/28-required-template-boost'
df = pd.read_csv(outdir / 'required_template_boost.csv')
df
```

```python
pivot = df.pivot(index='target_frac', columns='g_GeV_inv', values='required_boost').sort_index()
pivot
```

```python
fig, ax = plt.subplots(figsize=(7, 4.5))
for g, sub in df.groupby('g_GeV_inv'):
    ax.plot(sub['target_frac'], sub['required_boost'], marker='o', label=f'g = {g:.1e} GeV^-1')
ax.set_xscale('log')
ax.set_yscale('log')
ax.set_xlabel('Target fraction of anisotropic-CB limit')
ax.set_ylabel('Required template boost')
ax.set_title('Required normalization boost from current best point')
ax.grid(True, which='both', alpha=0.3)
ax.legend()
plt.show()
```
