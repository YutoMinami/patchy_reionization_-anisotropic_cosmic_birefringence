# 27b visualize natural-unit budget near $m_{\rm res}$

`27-natural_unit_budget_mres.py` の CSV を読み、visibility-weighted な patchy contribution が anisotropic-CB limit の何割を使うかを見る。


```python
from pathlib import Path
import pandas as pd
import matplotlib.pyplot as plt
from IPython.display import display

ROOT = Path('/home/yminami/workdir/CosmicBirefringence/PatchyReionization')
CSV_SUM = ROOT / 'results' / '27-natural-unit-budget-mres' / 'natural_unit_budget_mres_summary.csv'
CSV_SCAN = ROOT / 'results' / '27-natural-unit-budget-mres' / 'natural_unit_budget_mres_scan.csv'
summary = pd.read_csv(CSV_SUM).sort_values(["g_GeV_inv", "mass_over_mres"]).reset_index(drop=True)
scan = pd.read_csv(CSV_SCAN).sort_values(["g_GeV_inv", "mass_over_mres", "L"]).reset_index(drop=True)
display(summary.head())
summary.tail()
```

```python
best = summary.sort_values("max_frac_eff", ascending=False).head(6)
display(best)
```

```python
fig, ax = plt.subplots(figsize=(8, 4.8))
for g, sub in summary.groupby("g_GeV_inv"):
    ax.plot(sub["mass_over_mres"], sub["max_frac_eff"], marker="o", label=f"g={g:.1e}")
ax.axvline(1.0, color="C3", ls="--", lw=1.5, label="m/mres = 1")
ax.set_xlabel("mass / mres")
ax.set_ylabel("max D_tau_eff / D_limit")
ax.set_title("Visibility-weighted patchy budget near mres")
ax.grid(True, alpha=0.3)
ax.legend()
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(8, 4.8))
for g, sub in summary.groupby("g_GeV_inv"):
    ax.plot(sub["mass_over_mres"], sub["A_tau_eff_physical"].abs(), marker="o", label=f"|A_tau_eff|, g={g:.1e}")
ax.axvline(1.0, color="C3", ls="--", lw=1.5)
ax.set_yscale("log")
ax.set_xlabel("mass / mres")
ax.set_ylabel("|A_tau_eff physical|")
ax.set_title("Physical patchy coefficient near mres")
ax.grid(True, alpha=0.3)
ax.legend()
plt.show()
```

```python
row = best.iloc[0]
sub = scan[(scan["g_GeV_inv"] == row["g_GeV_inv"]) & (scan["mass_eV"] == row["mass_eV"])].copy()
fig, ax = plt.subplots(figsize=(8, 4.8))
ax.plot(sub["L"], sub["D_alphaalpha_tau_eff"], lw=2, label="patchy eff")
ax.plot(sub["L"], sub["D_alphaalpha_phi_bench"], lw=2, color="gray", label="genuine bench")
ax.plot(sub["L"], sub["D_alphaalpha_total_eff"], lw=2, label="total eff")
ax.plot(sub["L"], sub["D_alphaalpha_limit"], lw=2, ls="--", color="black", label="limit")
ax.set_xlabel("L")
ax.set_ylabel("D_L")
ax.set_title(f"Best visibility-weighted budget: g={row.g_GeV_inv:.1e}, m/mres={row.mass_over_mres:.3f}")
ax.grid(True, alpha=0.3)
ax.legend()
plt.show()
```
