# 13b Visualize A_CB Reinterpretation

`13-acb_reinterpretation_with_matched.py` の保存結果を使って、

- model-independent な anisotropic CB budget
- toy $C_L^{\tau\tau}$ template を置いたときの benchmark-dependent な制限

を図で読む notebook です。
        


```python
from pathlib import Path
import csv
from collections import defaultdict

import matplotlib.pyplot as plt
import numpy as np

ROOT = Path.cwd().resolve().parent
SCAN_CSV = ROOT / "results/13-acb-reinterpretation/acb_reinterpretation_scan.csv"
SUMMARY_CSV = ROOT / "results/13-acb-reinterpretation/acb_reinterpretation_summary.csv"


def read_csv(path):
    with path.open() as f:
        return list(csv.DictReader(f))


scan_rows = read_csv(SCAN_CSV)
summary_rows = read_csv(SUMMARY_CSV)

grouped = defaultdict(list)
for row in scan_rows:
    grouped[(row["case_name"], row["mass_eV"])].append(row)

sorted(grouped)[:3]
        
```

## patchy budget fraction


```python
mass_values = sorted({row["mass_eV"] for row in scan_rows})
selected_mass = mass_values[0]

plt.figure(figsize=(8, 5.5))
plt.axhline(1.0, color="black", ls="--", lw=1.0)
for (case_name, mass), items in grouped.items():
    if mass != selected_mass:
        continue
    items = sorted(items, key=lambda r: int(r["L"]))
    L = np.array([int(r["L"]) for r in items])
    frac = np.array([float(r["patchy_budget_fraction"]) for r in items])
    plt.loglog(L, frac, label=case_name)

plt.xlabel(r"$L$")
plt.ylabel(r"patchy budget fraction")
plt.title(rf"Patchy budget fraction at $m_a={float(selected_mass):.2e}\,{{\rm eV}}$")
plt.grid(True, which="both", alpha=0.3)
plt.legend()
plt.tight_layout()
plt.show()
        
```

## allowed template rescaling


```python
plt.figure(figsize=(8, 5.5))
plt.axhline(1.0, color="black", ls="--", lw=1.0)
for (case_name, mass), items in grouped.items():
    if mass != selected_mass:
        continue
    items = sorted(items, key=lambda r: int(r["L"]))
    L = np.array([int(r["L"]) for r in items])
    scale_lim = np.array([float(r["template_scale_limit"]) for r in items])
    plt.loglog(L, scale_lim, label=case_name)

plt.xlabel(r"$L$")
plt.ylabel(r"allowed template rescaling")
plt.title(rf"Allowed $C_L^{{\tau\tau}}$ template scaling at $m_a={float(selected_mass):.2e}\,{{\rm eV}}$")
plt.grid(True, which="both", alpha=0.3)
plt.legend()
plt.tight_layout()
plt.show()
        
```
