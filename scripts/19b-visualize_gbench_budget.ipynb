Chandra の `$g_{a\gamma\gamma}$` 制限を入れたときに、raw と physical の違いがどう見えるかを確認する notebook です。

```python
from pathlib import Path
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.image as mpimg

ROOT = Path('/home/yminami/workdir/CosmicBirefringence/PatchyReionization')
RESULT = ROOT / 'results/19-gbench-budget'
df = pd.read_csv(RESULT / 'gbench_budget_scan.csv')
df.head()
```

## Reading guide

- raw と physical を混ぜないこと
- physical は `A_tau=(g/2) A_tau_raw` を入れた後の absolute budget
- ただし field normalization の単位整合は still a caveat
